```python
import numpy as np
import random as rand
import matplotlib.pyplot as plt
from ddc_machine_learning.ml.preprocessing.pca import \
    PCA, size, transpose
```

##### Obtain data

```python
a_x = 0.05
a_y= 10
matrix =  [[n*(1+a_x*(rand.random()-0.5)),4*n+ a_y*(rand.random()-0.5)] for n in range(20)]
size(matrix)
```

```python
matrix_cols = transpose(matrix)
```

## PCA

```python
# Plot origional data
plt.scatter(matrix_cols[0], matrix_cols[1])
plt.title('Origional Data')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
pca = PCA(n_components=1)
pca.fit(matrix)
```

```python
data_pca = pca.transform(matrix)
pca_cols = transpose(data_pca)
size(data_pca)
```

```python
# Plot the data
plt.scatter(pca_cols[0], [0]*len(pca_cols[0]))
plt.title('PCA data (1-D)')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

### Reverse the PCA process
Information will be lost

```python
pca_reverse = pca.reverse(data_pca)
pca_reverse_cols = transpose(pca_reverse)
size(pca_reverse)
```

```python
# Plot the data
plt.scatter(matrix_cols[0], matrix_cols[1], label='Origional Data')
plt.scatter(pca_reverse_cols[0], pca_reverse_cols[1], label='PCA Data')
plt.title('Comparing Data')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
```
